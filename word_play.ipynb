```python
import random
```

```python
WORDS = [
    "python", "analysis", "dashboard", "automation", "consultant",
    "business", "strategy", "visualization", "data", "insights"
]
def choose_word():
    return random.choice(WORDS)
```

```python
def display_word(word, guessed_letters):
    displayed = ""
    for letter in word:
        if letter in guessed_letters:
            displayed += letter + " "
        else:
            displayed += "_ "
    return displayed.strip()
```

```python
def play_game():
    word = choose_word()
    guessed_letters = []
    attempts = 6

    print("\nüéØ Welcome to the Word Guessing Game!")
    print("Guess the word one letter at a time.")
    print("You have", attempts, "attempts.\n")

    while attempts > 0:
        print("Word:", display_word(word, guessed_letters))
        print("Guessed Letters:", guessed_letters)
        print("Remaining Attempts:", attempts)

        guess = input("Enter a letter: ").lower()

        # Input validation
        if not guess.isalpha() or len(guess) != 1:
            print("‚ö†Ô∏è Please enter a single alphabet letter.\n")
            continue

        if guess in guessed_letters:
            print("‚ö†Ô∏è You already guessed that letter.\n")
            continue

        guessed_letters.append(guess)

        if guess in word:
            print("‚úÖ Good guess!\n")
        else:
            attempts -= 1
            print("‚ùå Wrong guess!\n")

        # Check win condition
        if all(letter in guessed_letters for letter in word):
            print("üéâ Congratulations! You guessed the word:", word)
            return

    # If attempts finish
    print("üòû Game Over! The word was:", word)
```

```python
def main():
    while True:
        play_game()
        replay = input("\nDo you want to play again? (yes/no): ").lower()
        if replay != "yes":
            print("üëã Thanks for playing. Goodbye!")
            break

# -----------------------------
# RUN PROGRAM
# -----------------------------
if __name__ == "__main__":
    main()
```
